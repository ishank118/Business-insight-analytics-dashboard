```python
def get_second_debit(group):
    # Filter for all rows after the first one that are debits
    debits_after_first = group.iloc[1:][group.iloc[1:]['transaction_type'] == 'debit']
    if not debits_after_first.empty:
        return debits_after_first['category_name'].iloc[0]
    return None
merg['first_debit'] = merg.groupby('account_id', group_keys=False).apply(get_second_debit)
```

```python
month_users = (
    user_month
    .groupby('source_year_month')['account_id']
    .apply(set)
    .to_dict()
)
```

```python

segment_month_users = segment_month_users.copy()
segment_month_users['source_year_month'] = (
    pd.to_datetime(segment_month_users['source_year_month'].astype(str)).dt.to_period('M').astype(str)
)

months = sorted(segment_month_users['source_year_month'].unique())
segments = ['fast_active', 'medium_active', 'slow_active']

retention_rows = []

for i, month in enumerate(months):
    for segment in segments:
        # current month users for this segment
        curr_users = set(
            segment_month_users.loc[
                (segment_month_users['source_year_month'] == month) &
                (segment_month_users['active_segment'] == segment),
                'users'
            ].sum()
        )

        if i == 0:
            retained_users = None
            retention_rate = None
        else:
            prev_month = months[i - 1]

            prev_users = set(
                segment_month_users.loc[
                    (segment_month_users['source_year_month'] == prev_month) &
                    (segment_month_users['active_segment'] == segment),
                    'users'
                ].sum()
            )

            retained_users = len(prev_users & curr_users)
            retention_rate = round(retained_users / len(prev_users) * 100, 2) if prev_users else None

        retention_rows.append({
            "month": month,
            "segment": segment,
            "active_users": len(curr_users),
            "retained_users": retained_users,
            "retention_rate_pct": retention_rate
        })

retention_data = pd.DataFrame(retention_rows)
```

```python
user_metrics = merg[merg['transaction_type'] == 'debit'].groupby('account_id').agg({
    'active_segment': 'first',
    'txn_id': 'count',
    'amount': 'sum',
    'source_year_month': ['min', 'max']
}).reset_index()
```

```python
user_metrics.columns = ['account_id', 'segment', 'lifetime_transactions', 
                        'lifetime_amount', 'first_month', 'last_month']
```

```python
user_metrics.head()
```

```python
user_metrics['first_month'] = pd.to_datetime(user_metrics['first_month'])
user_metrics['last_month'] = pd.to_datetime(user_metrics['last_month'])
user_metrics['active_months'] = (
    (user_metrics['last_month'] - user_metrics['first_month']) / pd.Timedelta(days=30)
).round(1)
```

```python
user_metrics['active_months'] = user_metrics['active_months'].replace(0, 1)
```

```python
user_metrics['avg_monthly_transactions'] = (
    user_metrics['lifetime_transactions'] / user_metrics['active_months']
).round(2)
```

```python
user_metrics['avg_transaction_amount'] = (
    user_metrics['lifetime_amount'] / user_metrics['lifetime_transactions']
).round(2)
```

```python
user_metrics
```

```python
metrics_to_test = ['lifetime_transactions', 'avg_monthly_transactions', 
                   'active_months', 'avg_transaction_amount']
for metric in metrics_to_test:
    stats = user_metrics.groupby('segment')[metric].agg([
        ('Count', 'count'),
        ('Mean', 'mean'),
        ('Median', 'median'),
        ('Std Dev', 'std'),
        ('Min', 'min'),
        ('Max', 'max')
    ]).round(2)
```

```python
def hypothesis_test(df, metric, segment_col='segment'):
    # Prepare groups
    groups = {
        seg: df[df[segment_col] == seg][metric].dropna()
        for seg in df[segment_col].unique()
    }
    print(f"\nMetric: {metric}")
    for seg, vals in groups.items():
        print(f"{seg}: n = {len(vals)}")

    f_stat, p_anova = f_oneway(*groups.values())
    h_stat, p_kw = kruskal(*groups.values())
    grand_mean = df[metric].mean()
    ss_total = ((df[metric] - grand_mean) ** 2).sum()
    ss_between = sum(len(v) * (v.mean() - grand_mean) ** 2 for v in groups.values())
    eta_sq = ss_between / ss_total
    print(f"\nANOVA:            F = {f_stat:.2f}, p = {p_anova:.4f}")
    print(f"Kruskal-Wallis:   H = {h_stat:.2f}, p = {p_kw:.4f}")
    print(f"Effect Size :   {eta_sq:.3f}")

    return f_stat, p_anova, h_stat, p_kw, eta_sq
```

```python
for metric in metrics_to_test:
    hypothesis_test(user_metrics, metric)
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Banking.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
bins = [0,100000,300000,float('inf')]
labels = ['low_level','mid_level','high_level']
data['income_band'] = pd.cut(data['Estimated Income'],bins = bins,labels =labels)
```

```python
data['income_band'].value_counts().plot(kind = 'bar')
```

```python
data.head()
```

```python
cate = data[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification',
             'Properties Owned','Risk Weighting','IAId','income_band']]
for col in cate:
    display(data[col].value_counts())
```

```python
data
```

```python
for i, predictor in enumerate(data[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification',
             'Properties Owned','Risk Weighting','IAId','income_band']]):
    plt.figure(i)
    sns.countplot(data = data,x= predictor)
    
```

```python
for i, predictor in enumerate(data[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification',
             'Properties Owned','Risk Weighting','IAId','income_band']]):
    plt.figure(i)
    sns.countplot(data = data,x= predictor,hue = 'income_band')
```

```python
for i, predictor in enumerate(data[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification',
             'Properties Owned','Risk Weighting','IAId','income_band']]):
    plt.figure(i)
    sns.countplot(data = data,x= predictor,hue='GenderId')
```

```python
category = data[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification',
             'Properties Owned','Risk Weighting','IAId','income_band']]
for col in category:
    if col == 'Occupation':
        continue
    plt.figure(figsize = (8,10))
    sns.histplot(data[col])
    plt.title('histogram for Occupation')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.show()
```

### Numerical analysis 

```python
data.info()
```

```python
numerical = ['Estimated Income','Bank Loans','Superannuation Savings','Bank Deposits','Checking Accounts','Saving Accounts'
                  ,'Foreign Currency Account','Business Lending']
for i, col in enumerate (numerical):
    plt.figure(i)
    
    sns.histplot(data[col],kde = True)
    plt.title(col)
```

```python
correlation = data[numerical].corr()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(correlation,annot = True,cmap = 'crest',fmt ='.2f')
```
